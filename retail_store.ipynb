```python
# Подгружаем библиотеки
import pandas as pd
```

```python
# Считываем данные
retail_store = pd.read_csv('/mnt/HC_Volume_18315164/home-jupyter/jupyter-z-ravil/for_Dashbord/data1.csv', encoding='windows-1251')
```

```python
# Смотрим данные
retail_store
```

```python
# Данные в колонке 'InvoiceDate' приводим к типу datetime
retail_store['InvoiceDate'] = pd.to_datetime(retail_store['InvoiceDate'])
```

```python
# Провереям на пропущенные значения
retail_store.isna().sum()
```

```python
retail_store_new = retail_store.dropna()
```

```python
retail_store_new 
```

```python
# Удаляем строки с данными, где цены и количество меньше или равно нулю
retail_store_clean = retail_store_new.query('Quantity > 0 and UnitPrice > 0')
```

```python
# Записываем файл для создания дашборда
retail_store_clean.to_csv('retail_store_clean', index=False)
```

```python
# Для одного чатра на дашборде понадобился дополнительный график, для него сделали новый датасет
product_popular = retail_store_clean.groupby('Description', as_index=False).agg({'Quantity':'sum'})\
    .sort_values(by='Quantity', ascending=False).head(6)
```

```python
product_popular
```

```python
product_popular.to_csv('product_popular', index=False)
```
